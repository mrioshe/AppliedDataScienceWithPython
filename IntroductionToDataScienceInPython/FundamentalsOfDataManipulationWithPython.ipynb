### Data files and Summary statistics

```python
import csv
%precision 2

with open('mpg.csv') as csvfile:
    mpg=list(csv.DictReader(csvfile))

mpg[:3]
```

```python
len(mpg)
```

```python
mpg[0].keys()
```

```python
sum(float(d['cty']) for d in mpg)/len(mpg)
```

```python
sum(float(d['hwy']) for d in mpg)/len(mpg)
```

```python
cylinders=set(d['cyl'] for d in mpg)
cylinders
```

```python
CtyMpgByCyl=[]

for c in cylinders:
    summpg=0
    cyltypecount=0
    for d in mpg:
        if d['cyl']== c:
            summpg+=float(d['cty'])
            cyltypecount += 1
    CtyMpgByCyl.append((c,summpg/cyltypecount))

CtyMpgByCyl.sort(key=lambda x:x[0])
CtyMpgByCyl
```

```python
vehicleclass =set(d['class'] for d in mpg)
vehicleclass
```

```python
HwyMpgByClass=[]

for t in vehicleclass:
    summpg=0
    vclasscount=0
    for d in mpg:
        if d['class']==t:
            summpg +=float(d['hwy'])
            vclasscount +=1
    HwyMpgByClass.append((t,summpg/vclasscount))

HwyMpgByClass.sort(key=lambda x:x[1])
HwyMpgByClass
```

# Regex

Regular expressions, often abbreviated as regex or regexp, are powerful tools for matching patterns in text. In Python, the re module provides support for regular expressions. Here's a simple explanation:

Regular Expression:
A regular expression is a sequence of characters that defines a search pattern. It consists of literal characters (like letters and digits) and special characters (metacharacters) that have special meanings.

```python
import re
```

```python
#Match() checks for a match, that is at the beginning of the string and returns a boolean. Similarly, seatch(), checks for a match anywhere in the string, and 
#returns a boolean.

text = "Thos is a good day"

if re.search("good",text):
    print("wonderful!")
```

```python
# The findall() and split() functions will parse athe string for us and return chunks. 

text= "Amy works diligently. Amy gets good grades. Our student Amy is succesful"
re.split("Amy",text)
```

```python
len(re.findall("Amy",text))
```

The regex specification standard defines a markup language to describe patterns in text. lets start with anchors. Anchors specify the start and/or the end of the string that you are trying to match. The character ^ means start and the dollar sing character $ means end. If you put ^ before a string, it means that the text the regex processor retrieves must start with the string you specify. For ending, you have to put the $ character after the string, it means that the text Regex retrieves must end with the string you specify

```python
text= "Amy works diligently. Amy gets good grades. Our student Amy is succesful"
re.search("^Amy",text)
```

### Patterns and character classes

```python
# if we wanted to count the number of A's or B's in the list, we can't use "AB" since this is used to match all A's followed immediately by a B. Instead, we put 
# the characters A and B inside square brackets
re.findall("[AB]", grades)
```

```python
# Lets build a simple regex to parse out all instances where this student receive an A followed by a B or a C
re.findall("[A][B-C]", grades)

# You can write this pattern by using the pipe operator, witch means OR:

re.findall("AB|AC", grades)
```

### Quantifiers

Quantifiers are the number of tiumes you want a pattern to be matched in order to match. The most basic quantifier is expreseed as e{m,n}, where e is the expression or character we are matching, m is the minimum number of times you want it to matched, and n is the maximum number of times the item could be matched

```python
re.findall("A{2,10}",grades) #we'll use 2 as our min, but ten as our max
```

```python
re.findall("A{1,1}A{1,1}",grades)
#the first pattern is looking for any combination of two A's up to ten A's in a row. So ot sees four A's as a single streak. The second pattern is looking for 
# two A's back to back
```

```python
re.findall("A{2, 2}",grades)
```

```python
print(re.findall("A{2}",grades))
print(re.findall("AA",grades))
```

```python
# We could find a decreasing trend in a student's grades
re.findall("A{1,10}B{1,10}C{1,10}",grades)
```

```python
re.findall("AB{1,10}C{1,10}",grades)
```

### Groups

```python
re.findall("([\w ]*)(\[edit\])",wiki)
```

```python
for item in re.finditer("([\w ]*)(\[edit\])",wiki):
    print(item.groups())
```

```python
for item in re.finditer("([\w ]*)(\[edit\])",wiki):
    print(item.group(1))
```

```python
for item in re.finditer("(?P<title>[\w ]*)(?P<edit_link>\[edit\])",wiki):
    print(item.groupdict()['title'])
```

```python
print(item.groupdict())
```

We have seen how we can match individual character patterns with []. how we can group matches together using (), and how we can use quantifiers such as *,? or m{} to describe patterns. 

```python
# There are a number of short hands which are used with regexes for different kinds of characters, including:
# a . for any singles character wich is not a newline
# a \d for any digit
# and \s for any whitespace character, like spaces and tabs
# There are more, and a full list can be found in the python documentation for regexes
```

### Look-ahead and Look-behid

```python
for item in re.finditer("(?P<title>[\w ]+)(?=\[edit\])",wiki):
    print(item)
```

```python
regex=r"((https?://(www\.)?)|www\.)\w{3,}\.[a-z]{2,3}"
findall=re.match(regex, "www.www.moure10.dev")
print(findall)
```

```python
regex=r"^((https?://(www\.)?)|www\.)\w{3,}\.[a-z]{2,3}$"
findall=re.match(regex, "www.www.moure10.dev")
print(findall)
```

```python
regex=r"^((https?://(www\.)?)|www\.)\w{3,}\.[a-z]{2,3}$"
findall=re.match(regex, "www.moure10.dev")
print(findall)
```

```python
regex=r"^((https?://(www\.)?)|www\.)\w{3,}\.[a-z]{2,3}$"
findall=re.match(regex, "www.moure10.devskskskskss")
print(findall)
```

```python
regex=r"^((https?://(www\.)?)|www\.)\w{3,}\.[a-z]{2,3}$"
findall=re.findall(regex, "www.moure10.dev")
print(findall)
```
